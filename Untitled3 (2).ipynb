Simple Linear Regression (SLR) models the relationship between a predictor variable x.

```python
#question 2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Step 1: Simulate Data
np.random.seed(42)
n = 100  # number of data points
X = np.random.uniform(0, 10, n)  # Predictor variable X

# True parameters
beta_0 = 5  # Intercept
beta_1 = 2  # Slope
sigma = 3  # Standard deviation of the noise

# Generate error term (epsilon)
epsilon = np.random.normal(0, sigma, n)

# Generate outcome variable Y
Y = beta_0 + beta_1 * X + epsilon

# Step 2: Create a DataFrame with the data
data = pd.DataFrame({'X': X, 'Y': Y})

# Step 3: Fit the Simple Linear Regression model using statsmodels
model = smf.ols(formula='Y ~ X', data=data).fit()

# Step 4: Print the summary of the fitted model
print(model.summary())

# Step 5: Visualize the data and the fitted regression line
plt.scatter(data['X'], data['Y'], label='Observed Data', color='blue')
plt.plot(data['X'], model.fittedvalues, label='Fitted Line', color='red', linewidth=2)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression: Fitting the Line')
plt.legend()
plt.show()
```

```python
# question 3
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Step 1: Simulate Data
np.random.seed(42)
n = 100  # number of data points
X = np.random.uniform(0, 10, n)  # Predictor variable X

# True parameters
beta_0 = 5  # Intercept
beta_1 = 2  # Slope
sigma = 3  # Standard deviation of the noise

# Generate error term (epsilon)
epsilon = np.random.normal(0, sigma, n)

# Generate outcome variable Y
Y = beta_0 + beta_1 * X + epsilon

# Step 2: Create a DataFrame with the data
data = pd.DataFrame({'X': X, 'Y': Y})

# Step 3: Fit the Simple Linear Regression model using statsmodels
model = smf.ols(formula='Y ~ X', data=data).fit()

# Step 4: Visualize the data and both the true and fitted regression lines
plt.scatter(data['X'], data['Y'], label='Observed Data', color='blue')

# True regression line
plt.plot(X, beta_0 + beta_1 * X, label='True Line: Y = 5 + 2X', color='green', linestyle='--', linewidth=2)

# Fitted regression line from the model
plt.plot(data['X'], model.fittedvalues, label='Fitted Line', color='red', linewidth=2)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression: True vs Fitted Line')
plt.legend()
plt.show()
```

```python
# Import necessary libraries
import plotly.express as px
import statsmodels.api as sm
import pandas as pd

# Load the "Classic" Old Faithful Geyser dataset from Plotly's built-in data
old_faithful = px.data.gapminder().query("country == 'Old Faithful'")  # Placeholder, replace with correct data if available
# Alternatively, if using custom data, provide the data directly here.

# Create a scatter plot with a Simple Linear Regression trendline
fig = px.scatter(old_faithful, x='waiting', y='duration', 
                 title="Old Faithful Geyser Eruptions", 
                 trendline='ols')

# Apply LOWESS smoothing to the scatter plot
lowess = sm.nonparametric.lowess  # Local regression smoothing method
smoothed = lowess(old_faithful['duration'], old_faithful['waiting'], frac=0.25)  # Adjust `frac` as needed
smoothed_df = pd.DataFrame(smoothed, columns=['waiting', 'smoothed_duration'])

# Add LOWESS trendline to the scatter plot
fig.add_scatter(x=smoothed_df['waiting'], y=smoothed_df['smoothed_duration'], 
                mode='lines', name='LOWESS Trendline')

# Show the plot
fig.show()
```

The fitted values in a simple linear regression model are derived from the estimated coefficients  obtained after fitting the model. These coefficients, which represent the intercept and slope of the regression line, can be found in `fitted_model.params`. The fitted values are calculated by plugging the predictor variable values  X  into the regression equation y = beta0 + beta1*x, where beta0 and beta1 are the estimated parameters. In the model summary, `fitted_model.summary().tables[1]` displays these estimated coefficients along with their statistical significance. The `fitted_model.fittedvalues` contain the predicted Y-values, representing the model's best estimate of the true relationship between X and Y. These predicted values are the result of applying the estimated coefficients to the actual data points.

```python
#question 11

import plotly.express as px

# Create a new column 'kind' based on waiting time
old_faithful['kind'] = ['short' if w < 68 else 'long' for w in old_faithful['waiting']]

import statsmodels.formula.api as smf
from IPython.display import display

# Fit the model with the indicator variable 'kind'
model = smf.ols('duration ~ C(kind, Treatment(reference="short"))', data=old_faithful).fit()

# Display the model summary
display(model.summary().tables[1])

# Create a box plot of duration by kind
fig = px.box(old_faithful, x='kind', y='duration', 
             title='Duration by Kind (Short vs. Long)', 
             category_orders={'kind': ['short', 'long']})

# Show the plot
fig.show()
```
